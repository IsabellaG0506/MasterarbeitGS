```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["font.size"] = 14
```

```python
Pro = 74
Kontra = 279
ax = sns.barplot(x=["Pro", "Kontra"], y=[Pro, Kontra])
for i in ax.containers:
    ax.bar_label(i,)
plt.ylim(0, 310)
plt.grid(axis="y")
plt.savefig("./plots/insgesamt_pro_vs_kon.png", dpi=320)
```

```python
kategorie_farben = {
    1: (255, 0, 0),
    2: (255, 192, 0),
    3: (146, 208, 80),
    4: (0, 176, 80),
    5: (0, 176, 240),
    6: (0, 112, 192),
    7: (112, 48, 160),
    8: (128, 96, 0) # Sonstige
}
kat_farben = np.array(list(kategorie_farben.values())) / 255

kat_namen = ["1", "2", "3", "4", "5", "6", "7", "Sonstige"]
kategorien = [24, 33, 31, 43, 80, 20, 72, 60]

ax = sns.barplot(x=kat_namen, y=kategorien, palette=kat_farben)
for i in ax.containers:
    ax.bar_label(i,)
plt.ylim(0, 90)
plt.grid(axis="y")
plt.xlabel("Kategorie")
plt.savefig("./plots/insgesamt_kategorie.png", dpi=320)
```

```python
df = {
    "Zeitung": ["RP", "RP", "Tagesspiegel", "Tagesspiegel", "Welt", "Welt", "Zeit", "Zeit", "Spiegel", "Spiegel"],
    "Auslegung": ["Pro", "Kontra", "Pro", "Kontra", "Pro", "Kontra", "Pro", "Kontra", "Pro", "Kontra"],
    "Anzahl": [28, 25, 21, 72, 1, 64, 15, 72, 9, 46]
}
ax = sns.barplot(data=pd.DataFrame(df), x="Zeitung", y="Anzahl", hue="Auslegung")
for i in ax.containers:
    ax.bar_label(i,)
plt.ylim(0, 80)
plt.grid(axis="y")
plt.xlabel("")
plt.savefig("./plots/insgesamt_pro_kon_zeitung.png", dpi=320)
```

```python
df = pd.read_csv("kategorie_anzahl_pro_zeitung.csv", sep=";")
plt.figure(figsize=(11, 4))
ax = sns.barplot(data=df, x="Zeitung", y="Anzahl", hue="Kategorie", palette=kat_farben)
for i in ax.containers:
    ax.bar_label(i,)
plt.ylim(0, 38)
plt.legend(loc=(1.01, 0.1))
plt.grid(axis="y")
plt.xlabel("")
plt.tight_layout()
plt.savefig("./plots/zeitung_kategorie.png", dpi=320)
```
